```python
def developer_cleaning(x):
    return x.replace("[", "").replace("'", "").replace("]", "")

df['developer'] = df['developer'].apply(developer_cleaning)
df['developer']
```

```python
def dataframe_cleaning(x):
    return str(x).replace("[", "").replace("'", "").replace("]", "")

df = df.applymap(dataframe_cleaning)
df
```
